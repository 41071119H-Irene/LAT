# 比較各縣市補習班與區域數量比率是否存在城鄉差距

### 先讀入相關資料，擷取需要的部分

```python
import pandas as pd

df = pd.read_csv('Tainan.csv')

df.head()
```

```python
temp = df[['縣市','行政區別', '文理類合計']]
temp
```

```python
Type = temp.groupby(['行政區別'])['文理類合計'].sum().reset_index()
```

```python
Type['縣市'] = '台南市'
Type
```

```python
dfm = pd.read_csv('mazhu.csv')
```

```python
mazhu = dfm[['行政區別', '文理類合計']].groupby(['行政區別'])['文理類合計'].sum().reset_index()
```

```python
mazhu['縣市'] = '馬祖縣'
```

```python
mazhu
```

```python
frames = [Type, mazhu]

result = pd.concat(frames)
```
